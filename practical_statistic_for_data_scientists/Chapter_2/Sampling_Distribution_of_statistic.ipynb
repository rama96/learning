```python
filepath = dir_path.joinpath('Training Data.csv')
df = pd.read_csv(filepath)
df.columns = [i.lower() for i in df.columns]
df
```

```python
print(df.dtypes)
df.describe()
```

```python
# Complete data
complete_data = pd.DataFrame({'income':df['income'] , 'type':'data'})

# Picking 1000 samples at random
sample_data = pd.DataFrame({'income':df['income'].sample(1000) , 'type':'sample_data'})

# Picking mean of samples of 5 1000 times
sample_data_05 = pd.DataFrame({'income': [df['income'].sample(5).mean() for _ in range(1000)] , 'type':'mean_5'})

# Picking mean of samples of 20 1000 times
sample_data_20 = pd.DataFrame({'income':df['income'].sample(20) , 'type':'mean_20'})

# Picking mean of samples of 20 1000 times
sample_data_50 = pd.DataFrame({'income':df['income'].sample(20) , 'type':'mean_50'})


results = pd.concat([complete_data , sample_data , sample_data_05 , sample_data_20 , sample_data_50])
```
